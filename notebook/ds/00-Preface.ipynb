# คำนำ

เอกสารประกอบการสอนรายวิชาโครงสร้างข้อมูล 254251 ถูกสร้างขึ้นจากสมุดภาษาไพทอน (python notebook) ซึ่งเป็นเครื่องมือที่ถูกใช้ในการผลิตสื่อการสอนภาษาไพทอนอย่างแพร่หลายในระดับสากลเนื่องจากมีคุณสมบัติในการให้ความรู้ผ่านเนื้อหาควบคู่กับการเขียนรหัสแบบโต้ตอบ จึงเป็นแรงผลักดันให้ผู้สอนพัฒนาเอกสารประกอบการสอนผ่านสมุดภาษาไพทอนเพื่อตอบสนองนโนบายเอดูเคชัน 4.0 ในการทีจะพัฒนาเอกสารประกอบการสอนที่มีความทันสมัย และเกิดผลสัมฤทธิ์ที่ขึ้นกับนิสิตเป็นหลัก ดังนั้นหากไม่เป็นไปตามรูปแบบเอกสารที่เป็นรูปเล่มทางผู้สอนขอความอนุเคราะห์และขออภัยไว้ ณ ที่นี้

เรื่องการใช้ภาษาผู้สอนมีเจตนาที่จะใช้ตามหลักของราชบัณฑิตฯ ในส่วนของภาษาอังกฤษคำอื่น เช่น list str queue ฯลฯ ล้วนเป็นการอ้างถึงชนิดข้อมูลหรือคลาสของภาษาไพทอน หรือคลาสที่สร้างขึ้น หาได้เป็นการใช้ภาษาอังกฤษแทรกในภาษาไทยไม่ นอกจากนี้สมุดภาษาไพทอนยังมีข้อจำกัดเกี่ยวกับการแสดงผลสูตรทางคณิตศาสตร์ที่ใช้ฟอนท์ต่างจากส่วนของเนื้อหา

ในแง่ของเนื้อหาส่วนใหญ่ได้ถูกแปล ดัดแปลง และเรียบเรียงจาก [interactivepython](http://interactivepython.org/runestone/static/pythonds/index.html) [1] ดังนั้นจึงมีการนำรูปมาใช้โดยทุกครั้งจะมีการอ้างอิงไปสู่เวบไซต์ต้นฉบับ การอ้างอิงถึงรูปและตารางในสมุดภาษาไพทอนจะแตกต่างจากเอกสารทั่วไปโดยได้ทำการกำหนดลิงก์เพื่อสามารถกดเข้าถึงรูปหรือตารางนั้นๆ ได้โดยตรง หนังสือประกอบรายวิชามีดังนี้คือ [2] และ [3]

ทั้งนี้ในส่วนของภาพถ่ายสมุดภาษาไพทอน และสไลด์ได้มีการนำขึ้นแสดงต่อสาธารณะที่ https://github.com/fomapavlov/254251_1_2017
