# บทที่ 14 การประยุกต์ใช้กราฟ
บทเรียนก่อนหน้านี้ได้กล่าวถึงการสร้างโครงสร้างข้อมูลชนิดต่างๆ ในบทนี้จะขอกล่าวถึงการประยุกต์ใช้โครงสร้างข้อมูลกราฟร่วมกับโครงสร้างข้อมูลอื่นๆ

## 14.1 จุดประสงค์
* เพื่อให้เข้าใจถึงการประยุกต์ใช้กราฟ
* เพื่อให้เข้าใจถึงการประยุกต์ใช้โครงสร้างข้อมูลอื่นๆ ร่วมกับกราฟ
* เพื่อให้เข้าใจปัญหาต้นไม้ทอดข้ามน้อยสุด
* เพื่อให้เข้าใจปัญหาเส้นทางสั้นสุด


## 14.2 ต้นไม้ทอดข้ามน้อยสุด
ต้นไม้ทอดข้ามน้อยสุด (minimum spanning tree) เป็นการหากราฟย่อยจากกราฟไม่มีทิศทางที่มีคุณสมบัติแบบต้นไม้คือไม่มีการเกิดวงวน แต่มีเงื่อนไขเพิ่มเติมคือทุกจุดยอดในกราฟจะต้องอยู่ในต้นไม้และผลรวมของน้ำหนักหรือระยะทางของทุกเส้นเชื่อมรวมกันแล้วจะต้องมีค่าน้อยที่สุด

เพื่อลดความสับสนขอให้นิยามของต้นไม้ทอดข้ามโดยยังไม่สนใจเงื่อนไขน้อยสุดดังนี้

ต้นไม้ทอดข้ามของกราฟ $G = (V,E)$ มีคุณสมบัติดังนี้
* เป็นกราฟย่อยที่มีความเชื่อมโยงกันเป็นองค์ประกอบเดียว
* เป็นกราฟที่ไม่เกิดวงวน
* เป็นกราฟที่ประกอบด้วยทุกจุดยอดใน $V$

การหาต้นไม้ทอดข้ามน้อยสุดสามารถทำได้หลายวิธี ณ ทีนี้จะขอเสนอวิธีของ Prim (Prim's algorithm) ซึ่งมีรายละเอียดขั้นตอนวิธีดังนี้
* กำหนดให้ต้นไม้ที่ต้องการสร้างคือ $S$ เซตของเส้นเชื่อมคือ $A$
* เพิ่มจุดยอดเริ่มต้นของกราฟ $G$ เข้าไปใน $S$
* ตราบเท่าที่ $S$ ยังไม่ครอบคลุม $V$ ($V$ \ $S$ ไม่เป็นเซตว่าง)
    * ขยายต้นไม้โดยเลือกเส้นเชื่อม $e$ ที่มีค่าน้อยที่สุดที่มีเส้นเชื่อมกับ $S$
    * เพิ่มจุดยอดปลายทางของเส้นเชื่อม $e$ เข้าไปใน $S$
    * เพิ่มเส้นเชื่อม $e$ เข้าไปใน $A$
    
จากขั้นตอนวิธีส่วนขยายต้นไม้จะเห็นได้ว่ามีการเลือกเส้นเชื่อมที่มีค่าน้อยที่สุดและทุกๆ ครั้งที่มีการเพิ่มจุดยอดเข้าไปใน $S$ เส้นเชื่อมที่จะถูกพิจารณาจะมีจำนวนเพิ่มขึ้น เพื่อที่จะให้กระบวนการดังกล่าวมีประสิทธิภาพจึงจำเป็นต้องเลือกโครงสร้างข้อมูลที่เหมาะสม ซึ่งในกรณีนี้คือแถวคอยแบบมีความสำคัญหรือฮีพ ตัวอย่างของการสร้างต้นไม้ทอดข้ามน้อยสุดแสดงดัง [รูปที่ 14-3](#figure_03)

## 14.3 เส้นทางสั้นสุด
เป็นปัญหาที่ต้องการค้นหาเส้นทางสั้นสุดจากจุดยอดเริ่มต้นไปยังจุดยอดอื่นๆ ที่เหลือในกราฟแบบมีทิศทาง สามารถแก้ปัญหาได้ด้วยวิธีของ Dijkstra (Dijkstra's algorithm) โดยใช้การแยกจุดยอดเป็น 2 ประเภทคือ จุดยอดที่รู้ระยะแล้ว เก็บไว้ในโครงสร้างข้อมูลชื่อ Visited กับจุดยอดที่ยังไม่รู้ระยะ เก็บไว้ในโครงสร้งาข้อมูลชื่อ Unvisited และอาศัยหลักการง่ายๆ คือ หยิบจุดยอด $b$ จาก Unvisited ที่มีระยะทางไปถึงสั้นที่สุด เพิ่ม $b$ เข้าไปใน Visited แล้วทำการปรับค่าระยะทางที่ไปถึง $b$

อธิบายโดยละเอียดได้ด้วยขั้นตอนวิธีดังนี้
* เริ่มต้นระยะทางของจุดยอดเริ่มต้นเป็น 0
* เริ่มต้นระยะทางที่จะไปถึงจุดยอดอื่นๆ เป็น $\infty$
* ตราบเท่าที่ Unvisited ยังไม่ว่าง
    * หยิบจุดยอด $b$ จาก Unvisited ที่มีระยะทางสั้นสุด
    * เพิ่ม $b$ เข้าไปยัง Visited
    * สำหรับทุกจุดยอด $a$ ที่เชื่อมต่อกับ $b$
        * ปรับระยะของ $a$ เป็นค่าน้อยสุดระหว่าง ระยะเดิมของ $a$ หรือ (ระยะของ $b$ + ระยะจาก $b$ ไปยัง $a$)
        * หาก (ระยะของ $b$ + ระยะจาก $b$ ไปยัง $a$) มีค่าน้อยกว่าจากเงื่อนไขด้านบน ให้ปรับระยะ $a$ ด้วย (ระยะของ $b$ + ระยะจาก $b$ ไปยัง $a$)
        
จากขั้นตอนวิธีจะเห็นได้ว่ามีการหยิบจุดยอด $b$ จาก Unvisited ที่มีระยะทางสั้นสุด ดังนั้นจึงต้องเลือกโครงสร้างข้อมูลที่เหมาะสมสำหรับ Unvisited ซึ่งแถวคอยแบบมีความสำคัญหรือฮีพจะถูกนำมาใช้เพื่อให้ได้ประสิทธิภาพสูงสุด โดย Visited อาจใช้รายการของค่าความจริงเพื่อบอกว่าจุดยอดใดเคยถูกนำมาพิจารณาแล้ว และหากต้องการทราบเส้นทางยังสามารถเพิ่มรายการเพื่อเก็บว่าจุดยอดนั้นๆ มีเส้นทางมาจากจุดยอดใด ตัวอย่างของการหาเส้นทางสั้นสุด แสดงดัง [รูปที่ 14-4](#figure_04) (จาก [CSE 326: Data Structures](https://courses.cs.washington.edu/courses/cse326/07au/lectures/lect22.pdf)) โดย Cost คือระยะและ Found by คือรายการเพื่อเก็บว่าจุดยอดนั้นๆ มีเส้นทางมาจากจุดยอดใด
