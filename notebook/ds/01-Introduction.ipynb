# บทที่ 1 บทนำ
พื้นฐานของวิทยาการคอมพิวเตอร์นั้นจำเป็นต้องเข้าใจในเรื่องของขั้นตอนวิธี (algorithm) ในการแก้ปัญหาซึ่งจะต้องประกอบด้วยขั้นตอนที่ชัดเจนและกระชับ อันจะสามารถแปลงเป็นรหัสคอมพิวเตอร์หรือเรียกสั้นๆ ว่า รหัส (code) ได้ ซึ่งรหัสคอมพิวเตอร์ที่จะใช้ในรายวิชานี้คือภาษาไพทอนรุ่น 3 (python 3)

## 1.1 จุดประสงค์
* เพื่อให้เห็นถึงความสำคัญของความเข้าใจพื้นฐานเกี่ยวกับรูปแบบข้อมูล
* เพื่อให้เข้าใจถึงพื้นฐานข้อมูลแบบชนิดข้อมูลนามธรรม (abstract data type) และโครงสร้างข้อมูล


## 1.2 ข้อมูล
การที่นักพัฒนาจะทำการพัฒนาโปรแกรมบนคอมพิวเตอร์จำเป็นเสมอที่จะต้องมีการเก็บข้อมูล ซึ่งข้อมูลบนคอมพิวเตอร์ทั้งหลายทั้งปวงจะถูกเก็บอยู่ในรูปแบบทวิภาค (binary) หรือที่เข้าใจกันแบบง่ายๆ คือ 0 กับ 1 เช่น ‭01101101‬ เป็นข้อมูลที่เก็บในหน่วยความจำ ซึ่งแปลงให้อยู่ในรูปฐาน 10 ได้ 109 แต่แท้จริงแล้วข้อมูลนี้สามารถถูกมองได้ว่าเป็น 6D ในรูปฐาน 16 หรือ ตัวอักษร m ในตารางแอสกี้ก็ได้ (ทดสอบได้จากโปรแกรม charmap ในระบบปฎิบัติการวินโดวส์) ดังนั้นแล้วในการที่จะอ้างถึงข้อมูลในหน่วยความจำจึงมีความจำเป็นที่จะต้องกำหนดความต้องการที่จะใช้ข้อมูลนั้นในรูปแบบ (type) ใด หรือกำหนดว่าเป็นรูปแบบข้อมูล (data type) ใด ซึ่งการกำหนดรูปแบบข้อมูลนั้นจะมาพร้อมกับตัวดำเนินการ (operator) ตัวดำเนินการพื้นฐานมีตัวอย่างเช่น บวก ลบ คูณ หาร แต่ก็อาจจะมีฟังก์ชั่นอื่นๆ ที่ใช้เฉพาะเจาะจงเฉพาะรูปแบบข้อมูลนั้นๆ เพิ่มเติมเข้ามา


ภาษาโปรแกรมแต่ละภาษามักจะมีรูปแบบข้อมูลที่เป็นพื้นฐาน (primitives) ซึ่งรูปแบบข้อมูลที่เป็นพื้นฐานสามารถแบ่งได้เป็น 2 ประเภทย่อยๆ คือ รูปแบบข้อมูลพื้นฐานซึ่งประกอบด้วยรูปแบบข้อมูลตัวเดียว เช่น เลขจำนวนเต็ม (integer หรือ int) หรือเลขจำนวนเต็ม (float หรือ double) อีกประเภทคือรูปแบบข้อมูลแบบซับซ้อน ซึ่งจะประกอบด้วยรูปแบบข้อมูลพื้นฐานหลายๆ ตัวหรือหลายๆ รูปแบบ หรือประกอบด้วยข้อมูลแบบซับซ้อนซ้อนเข้าไป ในภาษาไพทอนนั้นมีการจัดหารูปแบบข้อมูลแบบซับซ้อนให้ใช้งานได้อย่างสะดวก เช่น วัตถุ (object), สายอักขระ (string), รายการ (list) และ พจนานุกรม (dictionary) ซึ่งรูปแบบข้อมูลเหล่านี้สามารถเก็บข้อมูลได้หลากหลายทั้งในแง่ประเภทและจำนวน อย่างไรก็ดีสิ่งที่ภาษาไพทอนมีให้อาจจะยังไม่เพียงพอกับการแก้ปัญหาที่มีความซับซ้อน ผู้พัฒนาจึงต้องมีการสร้างรูปแบบข้อมูลขึ้นมาเอง (user-defined type) เพื่อใช้แก้ปัญหาได้อย่างเหมาะสม

การใช้งานสมุดภาษาไพทอนในส่วนที่เป็นรหัสให้กด shift \+ enter เพื่อให้ได้ผลลัพธ์ (output)

```python
print(2+3*4)
print((2+3)*4)
print(2**10)
print(6/3)
print(7/3)
print(7//3)
print(7%3)
print(3/6)
print(3//6)
print(3%6)
print(2**100)
```

สิ่งต่างๆ ที่ปรากฎข้างต้นคือการใช้งานเบื้องต้นของรูปแบบข้อมูลประเภท int หรือ float และจากตัวอย่างให้ลองตั้งคำถามดูว่าอะไรที่ปรากฎให้เห็นในรหัสและผลลัพธ์บ้าง
1. print คืออะไร
2. \+ \- \* / % คืออะไร
3. // คืออะไร
4. \*\* คืออะไร
5. ตัวดำเนินการไหนให้ผลลัพธ์ออกมาเป็น int หรือ float เหตุใดผลลัพธ์บางตัวจึงมีจุดทศนิยมแต่บางตัวไม่มี

```python
4 * 1.2
```

## 1.3 แบบชนิดข้อมูลนามธรรม
ในการแก้ปัญหาทางวิทยาการคอมพิวเตอร์นั้นมักจะนิยมใช้การมองแบบนามธรรมเพื่อให้เห็นภาพรวมก่อนว่าปัญหานั้นๆ สามารถแก้ปัญหาได้ด้วยขั้นตอนวิธีใด โดยยังไม่ต้องสนใจว่าขั้นตอนวิธีที่จะนำมาใช้แก้ปัญหานั้นต้องเขียนโปรแกรมอย่างไร ซึ่งการมองปัญหาในลักษณะนี้ยังถูกถ่ายทอดมาถึงการใช้รูปแบบข้อมูลเพื่อแก้ปัญหา ดังนั้นเพื่อลดรายละเอียดในการสร้างรูปแบบข้อมูล ความสามารถหรือตัวดำเนินการของรูปแบบข้อมูลจึงถูกพิจารณาก่อน หลังจากนั้นจึงพิจารณารายละเอียดการสร้างโปรแกรมเพื่อให้ได้รูปแบบข้อมูลที่มีตัวดำเนินการอย่างที่ต้องการ ซึ่งมุมมองลักษณะนี้มีชื่อเรียกว่า แบบชนิดข้อมูลนามธรรม (abstract data type หรือ ADT) 

กล่าวคือในมุมมองของนักพัฒนาหรือผู้เขียนโปรแกรมจะกำหนดส่วนของแบบชนิดข้อมูลนามธรรมเพื่อให้สะดวกในมุมมองของผู้ใช้งานได้ ยกตัวอย่างเช่น ไลบรารี (library) ที่นักพัฒนาสร้างไว้เพื่อแก้ปัญหาเฉพาะ น้อยครั้งที่ผู้ใช้งานจะเข้าไปพิจารณาถึงรหัสว่าแบบชนิดข้อมูลนามธรรมสร้างอย่างไร แต่มักจะมองว่าฟังก์ชันที่มีให้รับข้อมูลขาเข้าอย่างไร แล้วให้ผลลัพธ์อะไร ซึ่งแบบชนิดข้อมูลนามธรรมจะใช้หลักการแบบเดียวกันคือ ผู้ใช้งานรู้แค่เพียงว่าแบบชนิดข้อมูลนามธรรมนั้นมีการต่อประสาน (interface) อย่างไร และใช้ให้เป็นเพื่อสามารถนำแบบชนิดข้อมูลนามธรรมไปใช้ในการแก้ปัญหาได้
